```python
import pandas as pd
```

```python
data = pd.read_csv("data/IMDB Dataset.csv", encoding="ISO-8859-1")
data.head()
```

```python
import string
import re
string.punctuation

def remove_punctuation_and_lower(text: str):
    clean = re.compile('<.*?>')
    html_removed = re.sub(clean, '', text)
    punctuation_free = "".join([i for i in html_removed if i not in string.punctuation])
    lower = punctuation_free.lower()
    return lower

data['clean_msg']= data['review'].apply(lambda x:remove_punctuation_and_lower(x))
data.head()
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
def tokenize(text: str):
    return tokenizer.tokenize(text)

data['msg_tokenized'] = data['clean_msg'].apply(tokenize)
```

```python
data.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Define a function to remove stop words from a sentence 
def remove_stop_words(word_list: list): 
  
  # Use a list comprehension to remove stop words 
  filtered_words = [word for word in word_list if word not in stop_words] 
  
  # Join the filtered words back into a sentence 
  return filtered_words
data['msg_no_stop_words'] = data['msg_tokenized'].apply(remove_stop_words)

data.head()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def lemmatize(text):
    doc = nlp(' '.join(text))  # Join tokens back into a string
    return [token.lemma_ for token in doc]

data['msg_lemmatized'] = data['msg_tokenized'].apply(lemmatize)

data.head()
```
